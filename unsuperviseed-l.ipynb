# Unsupervised Learning

* Unsupervised learning ek aisa ML technique hai jisme hum data ko labels (output) ke bina model ko dete hain.

* Yani model ko sirf input data diya jata hai , aur wo khud apne aap patterns, groups, structure find karta hai.

* “Model ko teacher nahi diya jata. Data se wo khud samajhta hai ki data ke andar kya chal raha hai.”

## Why in AI/ML



* AI aur ML me unsupervised learning bohot important hai kyunki:

    - Real world me labels milna mushkil hota hai :

        - Bohot saara data available hota hai, par usko label karna time-consuming aur expensive hota hai.

        - Example:
            
            - 10 lakh customer transactions ko manually label kaun karega?

    - Data ke andar hidden structure samajhna hota hai:

        - Kabhi humara goal prediction nahi hota, goal hota hai data ko samajhna.

    - Dimensionality reduction :

        - Large features ko compress karke model ko efficient banana.

### Unsupervised Learning ka main aim kya hota hai?

* Patterns dhundhna

* Groups banana

* Data ko visualize karna

* Similarity samajhna

* Noise remove karna

## Types of Unsupervised Learning

#### 1 : Clustering

* Clustering me model similar data ko ek group (cluster) me daal deta hai.

* Real example: Customer Segmentation

    - Mall ke paas 20,000 customer data hai:

        - age

        - income

        - spending score

    - Mall manager chahata hai ki customer kis type ke groups me divide hote hain.

* Model clusters banata hai:

    - Cluster 1: low income + low spending

    - Cluster 2: high income + high spending

    - Cluster 3: students (low income but trendy spending)

### 2: Dimensionality Reduction

* Jab features bohot jyada ho jaye (jaise 1000 columns), to model slow ho jata hai.

* Isliye unsupervised learning feature count kam karta hai without losing important information.

* PCA (Principal Component Analysis)

    - Used in:

        - image compression

        - noise removal

        - speeding up ML models

### 3: Association Rule Mining

* Ye shopping basket analysis me use hota hai.

* Example:

    - Agar koi customer milk kharidta hai, to wo bread bhi leta hai.

* Model rule banata hai:

    - "Milk → Bread"

* E-commerce use case:

    - “Aapne milk kharida, log bread bhi kharidte hain.”
