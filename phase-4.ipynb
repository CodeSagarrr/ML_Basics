## 1 : Accuracy

* Accuracy simple shabdo mein yeh batata hai ki model ne kitne predictions sahi kiye.

* Formula :

    - Accuracy = (Correct Predictions) / (Total Predictions)

* Why we use in ML:

    - Beginner friendly metric

    - Directly batata hai model kitni baar sahi tha

    - Easy to understand

* Accuracy imbalance problem : 

    - Agar dataset mein ek class bahut zyada ho aur ek kam ho, accuracy misleading ho sakti hai.

* Example:

    - 980 normal, 20 fraud cases

    - Agar model sabko “normal” bol de:

    - Accuracy = 98%

    - But model useless hai → fraud detect hi nahi kiya.

```python
y_true = [ 1 , 0 , 1 , 1 , 0 ]

y_pred = [ 1 , 0 , 0 , 1 , 0 ]
```

* Correct predictions = 4/5

* Accuracy = 0.80

## 2 : F1 Score

* F1 score precision aur recall ka harmonic mean hota hai. ( Precision and recall are metrics used to evaluate the performance of a classification model )

* Useful jab classes imbalanced ho.

* F1 Score formula :

    - F1 = 2 * (precision * recall) / (precision + recall)

* Why we use in ML:

    - Jab dataset imbalanced ho, F1 score real performance batata hai

    - Precision aur Recall balance karta hai

    - Fraud detection / Health data / Spam detection mein bahut important

* Precision :

    - Model ne jo “positive” bola, unme se kitne sach mein positive hai.

* Recall :

    - Actual positive mein se model kitne detect kar paya.

## 3 : MSE – Mean Squared Error

* Yeh regression models ke liye metric hota hai (classification ke liye nahi).

* MSE batata hai ki predicted value aur real value ke beech average squared error kitna hai.

* Formula:

    - MSE = average( (y_true – y_pred)^2 )

* Why we use in ML:

    - Measures how far predictions are from real values

    - Squaring se big error ko zyada weight milta hai

    - House price, stock prediction, regression models mein use hota hai



```python
y_true = [5, 6, 7]

y_pred = [4, 6, 10]
```

* (5-4)^2 = 1

* (6-6)^2 = 0

* (7-10)^2 = 9

* MSE = (1 + 0 + 9) / 3 = 3.33

## 4 : Confusion Matrix

* Confusion matrix table hoti hai jo batati hai ki model ne kaunse predictions sahi ya galat kiye.

* Structure:

    - Predicted 0   ,  Predicted 1
    - Actual 0              - TN               FP
    - Actual 1              - FN               TP

* Why we use in ML:

    - Pure model performance ek table mein

    - Precision, Recall, F1 yahin se nikalte hain

    - Samajh aata hai model kis class mein mistake kar raha hai

* TP (True Positive) – correct positive

* TN (True Negative) – correct negative

* FP (False Positive) – false alarm

* FN (False Negative) – missed detection



```python
[[85, 20]
 [16, 58]]
```

* Means:

    - 85 actual negative → correctly predicted negative

    - 20 actual negative → galat se positive

    - 16 actual positive → galat se negative

    - 58 actual positive → correctly predicted positive
