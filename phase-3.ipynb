# Clustering

* Clustering ek unsupervised learning technique hai jisme machine similar data points ko ek group me rakhti hai jise cluster kehte hain.

* Isme koi target column nahi hota.

### 1 : Cluster

* Cluster ek group hota hai jisme data points ek jaisi properties rakhte hain.

* Jaise shopping habits, height weight, marks pattern.

### 2 : Similarity

* Similarity ka matlab ki data points ek dusre ke kitne close hain values ke hisaab se.

### 3 : Distance Measure

* Similarity ko measure karne ke liye distance calculate hota hai.

    - Most common: Euclidean distance

* Formula:

    - Distance = √((x2−x1)² + (y2−y1)²)

* Agar distance kam hai to points similar hain.

### Cluster Example

* Points:

    - (10, 20), (11, 22), (100, 90)

* Pehle do points ek dusre ke close hain . Third point in sabse door hai , Yani 2 clusters ban sakte hain.

# K Means 

* K Means ek popular clustering algorithm hai jisme hum data ko K clusters me divide karte hain. 

* har cluster ka ek centroid banata hai. Centroid cluster ka average point hota hai.

* K Means ka full meaning

    - K = number of clusters

    - Means = average (centroid) jiske around cluster banta hai.

* Algorithm do important cheeze karta hai

    - Random points ko centroid banata hai

    - Har data point ko closest centroid ke paas assign karta hai

    - Fir centroid ko update karta hai

    - Ye steps repeat hote hain jab tak cluster stable na ho jaye

### Centroid

* Centroid cluster ka center point hota hai.

* Agar cluster me points:

* (10,20), (12,22)

* Centroid = average = (11, 21)

```python
from sklearn.cluster import KMeans
import numpy as np

X = np.array([
    [10, 20],
    [12, 22],
    [90, 80],
    [85, 82]
])

model = KMeans(n_clusters=2) # n_cluster = how many cluster you wants basec on the data
model.fit(X)

print("Labels:", model.labels_)
print("Centroids:", model.cluster_centers_)
```

# Visualize Clusters

* Clusters ko visualize karne ka matlab hai graph me dekhna ki kaunse points kis group me gaye aur centroids kaha baney.


### 1 : Scatter Plot

* Clusters dikhane ke liye scatter plot use hota hai jisme points 2D graph me show hote hain.

### 2 : Color Coding

* Har cluster ko alag color me show kiya jata hai taaki clearly pata chale kaunse points kis cluster me hain.

### 3 : Centroid Display

* Centroid ko normal points se alag dikhaya jata hai

* mostly marker="X" use hota hai.

```python
import matplotlib.pyplot as plt

plt.scatter(X[: , 0] , X[: , 1] , c=model.labels_)
plt.scatter(model.cluster_centers_[:,0] , model.cluster_centers_[:,1] , s=200 , marker="X")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K Means Cluster Visualization")
plt.show()
```
